# Compressed Sensing Raw-data

```python
%matplotlib inline

import matplotlib.pylab as plt
import numpy as np
import os
import glob
import pickle
```

```python
# Train Set 
kspace_files = np.asarray(glob.glob("/media/roberto/DATA/GDrive/MR-Data/Normative-Raw-Data/Train/*.npy"))
indexes = np.arange(kspace_files.size,dtype = int)
np.random.shuffle(indexes)
kspace_files_train = kspace_files[indexes]

# Validation set
kspace_files_val = np.asarray(glob.glob("/media/roberto/DATA/GDrive/MR-Data/Normative-Raw-Data/Val/*.npy"))

# Test set
kspace_files_test = np.asarray(glob.glob("/media/roberto/DATA/GDrive/MR-Data/Normative-Raw-Data/Test/*.npy"))

print kspace_files_train[0]
print len(kspace_files_train)
print 
print kspace_files_val[0]
print len(kspace_files_val)
print 
print kspace_files_test[0]
print len(kspace_files_test)
```

## Sample Train k-space and image reconstruction

```python
sample_kspace_train = np.load(kspace_files_train[0])
sample_rec_train = np.abs(np.fft.ifft2(sample_kspace_train[:,:,:,0]+1j*sample_kspace_train[:,:,:,1]))
slice_index = 90
plt.figure()
plt.subplot(121)
plt.imshow(np.log(1+np.abs(sample_kspace_train[slice_index,:,:,0]+1j*sample_kspace_train[slice_index,:,:,1]))\
           ,cmap = 'gray')
plt.axis("off")
plt.subplot(122)
plt.imshow(sample_rec_train[slice_index],cmap = 'gray')
plt.axis("off")
plt.show()
```
